```python
######################Funções########################

# Importando a biblioteca Tkinter - ambiente gráfico do app:
from tkinter import *
import tkinter as tk
# Importando a biblioteca matplotlib - gerar o gráfico de barras do app:
import matplotlib.pyplot as plt
# Permite a criação de uma figura e plotar o gráfico dentro desta figura:
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
import pandas as pd

# Variável globarl 'var'
global var

# Função cálculo dos investimentos:
def calc_invest():
    
    # Variável pra criar uma lista com todos os investimentos após o cálculo:
    todos_invest = []
    
    # Variável globarl 'var'
    global var
    
    # Dados iniciais para o cálculo dos investimentos:
    var = float(entry11.get())
    acrescido_inicial = float(entry2.get())
    meses_inicial = float(entry3.get())
    juros_inicial = float(entry4.get())
            
    # Cálculo do valor bruto:
    bruto = var + (acrescido_inicial * (meses_inicial - 1))
    
    # Cálculo do valor na poupança:
    juros_poupanca = 0.51 / 100
    poupanca = (1 + juros_poupanca) * ((((1 + juros_poupanca)**meses_inicial)-1)/juros_poupanca) * acrescido_inicial
    
    # Limpar a caixa de texto a cada novo cálculo:
    entry0.delete(0, END)
    
    # Loop para calcular o CDI/CDB:
    for i in range(int(meses_inicial)):    
        montante = float(var * ((1+(juros_inicial/100))**1))
        var = float(montante + acrescido_inicial)

    # Inserir o resultado no campo 'Valor final do investimento':
    # O primeiro argumento é a posição inicial:
    # O segundo argumento é a informação que será impressa: 
    entry0.insert(0, f'{montante:_.2f}'.replace('.', ',').replace('_', '.'))
    
    # Dados de localização do 'entry0' na janela do app:
    entry0.place(
        x = 364.0, y = 202,
        width = 70.0,
        height = 21)
    
    # Adicionar os dados na lista 'todos invest':
    todos_invest.append(bruto)
    todos_invest.append(poupanca)
    todos_invest.append(montante)
    
    # Adicionando a lista 'todos_invest' a variável 'var':
    var = todos_invest
    
    # Chamando a função 'grafico()' para receber a variável/lista 'var':
    grafico()
        
# Função conversão juros:
def conv_juros():
    # Pegar a informação do campo %anual (entry6):
    entrada_juros = float(entry6.get())
    # Cálculo para transformar juros anual para mensal:
    taxa_mensal = float(entrada_juros / 12)
    #juros_mensal = entrada_juros 
    # Limpar o texto da caixa %Mensal a cada nova conversão:
    entry1.delete(0, END)
    # Inserir o resultado no campo %Mensal (entry6):
    # O primeiro argumento é a posição e o segundo é o texto (variável no caso): 
    entry1.insert(0, f'{taxa_mensal:,.2f}')
    # Informando a localização da resposta na janela app:
    entry1.place(
        x = 483.0, y = 288,
        width = 50.0,
        height = 21)
    
# Função conversão tempo:
def conv_tempo():
    # Pegar a informação do campo anos (entry7):
    entrada_anos = int(entry7.get())
    # Cálculo para transformar anos para meses:
    meses = int(entrada_anos * 12)
    # Limpar o texto da caixa Meses a cada nova conversão:
    entry5.delete(0, END)
    # Inserir o resultado no campo meses (entry8):
    # O primeiro argumento é a posição e o segundo é o texto (variável no caso): 
    entry5.insert(0, f'{meses}')
    # Informando a localização da resposta na janela app:
    entry5.place(
        x = 483.0, y = 342,
        width = 40.0,
        height = 21)
    
# Função Conversão do CDI:
def conv_cdi_cdb():
    cdi_anual = float(entry8.get())
    cdi_acrescimo = float(entry9.get())
    taxa_anual = (cdi_anual * (cdi_acrescimo / 100))
    taxa_mensal = (taxa_anual / 12)
    entry10.delete(0, END)
    # Inserir o resultado no campo %Mensal (entry10):
    # O primeiro argumento é a posição e o segundo é o texto (variável no caso): 
    entry10.insert(0, f'{taxa_mensal:,.2f}')
    # Informando a localização da resposta na janela app:
    entry10.place(
        x = 466.0, y = 438,
        width = 40.0,
        height = 21) 

# Função gráfico: plota o gráfico de barras na tela do app:
# É inicializada através da função 'calc_invest':
def grafico():
    global var
    print(var)

    # Criando a figura e o tamanho dela junto com o dpi:
    figura = plt.Figure(figsize = (3,2), dpi=50.0)
    ax = figura.add_subplot(111)
    # Após importar 'FigureCanvasTkAgg', configurar o canva para plotar o 
    # gráfico dentro do Tkinter:
    canva = FigureCanvasTkAgg(figura, window)
    # Posicionando o gráfico, largura e altura:
    canva.get_tk_widget().place(
        x = 130.0, y = 510,
        width = 405.0,
        height = 250.0)

    # Definindo os valores do eixo x e y:
    x = var
    y = ['Bruto', 'Poupança', 'CDI/CDB']

    # Título e tamanho da letra:
    ax.set_title('Investimentos', fontsize = 22)
    # Rótulo do eixo x e tamanho da fonte:
    ax.set_xlabel('Reais', fontsize = 17)
    # Tipo do gráfico e cores das barras:
    ax.barh(y, x, color = ['#3976e6', '#102382', '#16104B'])
    # Retirando as spines da direita e de cima do gráfico:
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
            
    # Configurando tamanho da letra, cor, rotação do rótulo do eixo 'X':
    for label in ax.yaxis.get_ticklabels():
        label.set_color('black')
        label.set_rotation(0)
        label.set_fontsize(12) 
        
    # Configurando tamanho da letra, cor, rotação do rótulo do eixo 'Y':
    for label in ax.xaxis.get_ticklabels():
        label.set_color('black')
        label.set_rotation(0)
        label.set_fontsize(15)   
        
    # Rótulos de valores das barras principais do gráfico:
    for i, eixo in enumerate(var):
        txt = f'R$ {eixo:,.2f}'.replace('.', ',').replace('_', '.')
        x_coord = eixo +0.2
        y_coord = i -0.1
        ax.text(y = y_coord, x = x_coord, s = txt, fontsize = 15, color = 'black')
      
        
######################Tkinter########################

# Atribuindo 'Tk()' a variável 'window':
window = Tk()

# Título da janela do Tkinter:
window.title('RAS Investimenros - Renda Fixa')

# Medidas da janela do app:
window.geometry("580x790")

# Atribuindo uma 'StringVar' a variável 'var':
var = tk.StringVar()

#Canvasonfiguração do Canvas: largura, altura, cor e outros:
window.configure(bg = "#ffffff")
canvas = Canvas(
    window,
    bg = "#ffffff",
    height = 790,
    width = 580,
    bd = 0,
    highlightthickness = 0,
    relief = "ridge")
canvas.place(x = 0, y = 0)

background_img = PhotoImage(file = f"background.png")
background = canvas.create_image(
    290.0, 245.0,
    image=background_img)

####################Front-end########################

entry0_img = PhotoImage(file = f"img_textBox0.png")
entry0_bg = canvas.create_image(
    448.5, 211.5,
    image = entry0_img)

entry0 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry0.place(
    x = 362.0, y = 200,
    width = 173.0,
    height = 21)

entry1_img = PhotoImage(file = f"img_textBox1.png")
entry1_bg = canvas.create_image(
    509.0, 298.5,
    image = entry1_img)

entry1 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry1.place(
    x = 483.0, y = 287,
    width = 52.0,
    height = 21)

entry2_img = PhotoImage(file = f"img_textBox2.png")
entry2_bg = canvas.create_image(
    448.5, 72.5,
    image = entry2_img)

entry2 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry2.place(
    x = 362.0, y = 61,
    width = 173.0,
    height = 21)

entry3_img = PhotoImage(file = f"img_textBox3.png")
entry3_bg = canvas.create_image(
    448.5, 105.5,
    image = entry3_img)

entry3 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry3.place(
    x = 362.0, y = 94,
    width = 173.0,
    height = 21)

entry4_img = PhotoImage(file = f"img_textBox4.png")
entry4_bg = canvas.create_image(
    448.5, 138.5,
    image = entry4_img)

entry4 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry4.place(
    x = 362.0, y = 127,
    width = 173.0,
    height = 21)

img0 = PhotoImage(file = f"img0.png")
b0 = Button(
    window,
    image = img0,
    borderwidth = 0,
    highlightthickness = 0,
    command = calc_invest,
    relief = "flat")

b0.place(
    x = 448, y = 163,
    width = 90,
    height = 24)

img1 = PhotoImage(file = f"img1.png")
b1 = Button(
    image = img1,
    borderwidth = 0,
    highlightthickness = 0,
    command = conv_juros,
    relief = "flat")

b1.place(
    x = 360, y = 287,
    width = 113,
    height = 23)

img2 = PhotoImage(file = f"img2.png")
b2 = Button(
    image = img2,
    borderwidth = 0,
    highlightthickness = 0,
    command = conv_tempo,
    relief = "flat")

b2.place(
    x = 360, y = 341,
    width = 113,
    height = 23)

entry5_img = PhotoImage(file = f"img_textBox5.png")
entry5_bg = canvas.create_image(
    509.0, 352.5,
    image = entry5_img)

entry5 = Entry(
    window,
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry5.place(
    x = 483.0, y = 341,
    width = 52.0,
    height = 21)

entry6_img = PhotoImage(file = f"img_textBox6.png")
entry6_bg = canvas.create_image(
    324.0, 298.5,
    image = entry6_img)

entry6 = Entry(
    window,
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry6.place(
    x = 298.0, y = 288,
    width = 52.0,
    height = 21)

entry7_img = PhotoImage(file = f"img_textBox7.png")
entry7_bg = canvas.create_image(
    324.0, 352.5,
    image = entry7_img)

entry7 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry7.place(
    x = 298.0, y = 342,
    width = 52.0,
    height = 21)

img3 = PhotoImage(file = f"img3.png")
b3 = Button(
    image = img3,
    borderwidth = 0,
    highlightthickness = 0,
    command = conv_cdi_cdb,
    relief = "flat")

b3.place(
    x = 342, y = 437,
    width = 113,
    height = 23)

entry8_img = PhotoImage(file = f"img_textBox8.png")
entry8_bg = canvas.create_image(
    165.0, 448.5,
    image = entry8_img)

entry8 = Entry(
    window,
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry8.place(
    x = 132.0, y = 437,
    width = 66.0,
    height = 21)

entry9_img = PhotoImage(file = f"img_textBox9.png")
entry9_bg = canvas.create_image(
    282.0, 448.5,
    image = entry9_img)

entry9 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry9.place(
    x = 232.0, y = 437,
    width = 100.0,
    height = 21)

entry10_img = PhotoImage(file = f"img_textBox10.png")
entry10_bg = canvas.create_image(
    500.0, 448.5,
    image = entry10_img)

entry10 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry10.place(
    x = 465.0, y = 437,
    width = 70.0,
    height = 21)

entry11_img = PhotoImage(file = f"img_textBox11.png")
entry11_bg = canvas.create_image(
    448.5, 39.5,
    image = entry11_img)

entry11 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry11.place(
    x = 362.0, y = 27,
    width = 173.0,
    height = 23)

window.resizable(False, False)
window.mainloop()

"""
Ajuda geral da calculadora do cidadão

A Calculadora do cidadão é uma aplicação interativa, de acesso público, que permite simular situações do cotidiano 
financeiro. Após a escolha do serviço financeiro, os cálculos são realizados a partir de informações fornecidas pelo 
usuário. No caso de correção de valores, também são utilizadas séries históricas de taxas e indicadores financeiros.

Serviços financeiros disponíveis para simulação:

Aplicação com depósitos regulares
Financiamento com prestações fixas
Valor futuro de um capital
Correção de valores
Aplicação com depósitos regulares
É a situação de aplicações mensais e de mesmo valor, considerando uma determinada taxa de juros, obtendo um valor ao 
final de um número de meses. Quatro informações estão envolvidas:
Nº de meses
Taxa de juros mensal
Valor do depósito regular
Valor obtido ao final
Informe 3 valores e pressione o botão [Calcular] para obter o 4°
Financiamento com prestações fixas
É a situação de pagamentos mensais e de mesmo valor, considerando uma determinada taxa de juros, liquidando um valor 
financiado ao final de um número de meses. O cálculo considera juros compostos e capitalização mensal.
Quatro informações estão envolvidas:
Nº de meses
Taxa de juros mensal
Valor da prestação
Valor financiado
Informe 3 valores e pressione o botão [Calcular] para obter o 4°
Obs.: O cálculo da taxa de juros é feito por aproximação do valor da prestação com margem de erro inferior a 0.000001.
Valor futuro de um capital
É a situação que um valor atual é projetado no futuro, considerando uma determinada taxa de juros, obtendo um valor ao 
fim de um número de meses.
Quatro informações estão envolvidas:
Nº de Meses
Taxa de Juros Mensal
Capital Atual
Valor Obtido ao Final
Informe 3 valores e pressione o botão [Calcular] para obter o 4°
Correção de valores

Índices de preços
A atualização é obtida multiplicando-se o valor a ser corrigido (ou 1 se não informado) pelo fator acumulado do índice 
de referência (Ex.: produtório dos índices mensais de IPCA/100+1). São usados no cálculo os índices da data inicial e 
da data final. Assim sendo, caso deseje a correção por apenas um mês, o usuário deve informar a data inicial igual à 
data final.

TR
Taxa Referencial. Taxa obtida a partir das médias dos CDBs de 30 dias a taxas pré-fixadas praticadas por bancos comerciais.

Poupança
Os rendimentos da poupança são creditados mensalmente, na data equivalente à data de aplicação (data-base). Dessa forma, 
se uma aplicação na poupança for resgatada antes de chegar à sua primeira data-base, não fará jus a qualquer correção, 
e o valor final será igual ao inicial.

Selic
Utiliza para correção a taxa apurada no Selic. Esta taxa é obtida mediante o cálculo da taxa média ponderada e ajustada 
das operações de financiamento por um dia, lastreadas em títulos públicos federais e cursadas no referido sistema ou em 
câmaras de compensação e liquidação de ativos, na forma de operações compromissadas.

CDI
Depósito Interfinanceiro. É uma operação realizada exclusivamente entre instituições financeiras, para permitir a troca 
de reservas bancárias entre elas. Tem registro no CETIP - Câmara de Custódia e Liquidação. Foi autorizada pelo 
Decreto-Lei nº 2.284/86 e regulamentada pela Resolução nº 1.647/89. A taxa do CDI é calculada como a média das 
taxas das operações registradas, ponderadas pelos volumes negociados, em determinado dia útil.

"""
```
