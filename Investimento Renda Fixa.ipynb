```python
from tkinter import *
import tkinter as tk
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
import pandas as pd

global valor_inicial

def calc_invest():
    
    todos_invest = []
    
    global valor_inicial
    
    # Dados iniciais para o cálculo dos investimentos:
    valor_inicial = float(entry11.get())
    acrescido_inicial = float(entry2.get())
    meses_inicial = float(entry3.get())
    juros_inicial = float(entry4.get())
            
    # Cálculo do valor bruto:
    bruto = valor_inicial + (acrescido_inicial * (meses_inicial - 1))
    
    # Cálculo do valor na poupança:
    juros_poupanca = 0.51 / 100
    poupanca = (1 + juros_poupanca) * ((((1 + juros_poupanca)**meses_inicial)-1)/juros_poupanca) * acrescido_inicial
    
    entry0.delete(0, END)
    
    # Calcular o CDI/CDB:
    for i in range(int(meses_inicial)):    
        montante = float(valor_inicial * ((1+(juros_inicial/100))**1))
        valor_inicial = float(montante + acrescido_inicial)

    # Inserir o resultado no campo 'Valor final do investimento':
    entry0.insert(0, f'{montante:_.2f}'.replace('.', ',').replace('_', '.'))
    entry0.place(
        x = 364.0, y = 202,
        width = 70.0,
        height = 21)
    
    # Adicionar os dados na lista 'todos invest':
    todos_invest.append(bruto)
    todos_invest.append(poupanca)
    todos_invest.append(montante)
    
    valor_inicial = todos_invest
    grafico()
        
# Função conversão juros:
def conv_juros():
    entrada_juros = float(entry6.get())
    taxa_mensal = float(entrada_juros / 12)
    entry1.delete(0, END)
    # Inserir o resultado no campo %Mensal (entry6):
    entry1.insert(0, f'{taxa_mensal:,.2f}')
    entry1.place(
        x = 483.0, y = 288,
        width = 50.0,
        height = 21)
    
# Função conversão tempo:
def conv_tempo():
    entrada_anos = int(entry7.get())
    meses = int(entrada_anos * 12)
    entry5.delete(0, END)
    entry5.insert(0, f'{meses}')
    entry5.place(
        x = 483.0, y = 342,
        width = 40.0,
        height = 21)
    
# Função Conversão do CDI:
def conv_cdi_cdb():
    cdi_anual = float(entry8.get())
    cdi_acrescimo = float(entry9.get())
    taxa_anual = (cdi_anual * (cdi_acrescimo / 100))
    taxa_mensal = (taxa_anual / 12)
    entry10.delete(0, END)
    entry10.insert(0, f'{taxa_mensal:,.2f}')
    entry10.place(
        x = 466.0, y = 438,
        width = 40.0,
        height = 21) 

# Função gráfico: plota o gráfico de barras na tela do app:
def grafico():
    global valor_inicial
    figura = plt.Figure(figsize = (3,2), dpi=60)
    ax = figura.add_subplot(111)
    canva = FigureCanvasTkAgg(figura, window)
    canva.get_tk_widget().place(
        x = 130.0, y = 510,
        width = 405.0,
        height = 250.0)

    y = valor_inicial
    x = ['Bruto', 'Poupança', 'CDI/CDB']

    ax.set_title('Investimentos', fontsize = 19)
    ax.set_ylabel('Reais', fontsize = 14)
    ax.bar(x, y, color = ['#3976e6', '#102382', '#16104B'])
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
            
    for label in ax.yaxis.get_ticklabels():
        label.set_color('black')
        label.set_rotation(0)
        label.set_fontsize(12) 
        
    for label in ax.xaxis.get_ticklabels():
        label.set_color('black')
        label.set_rotation(0)
        label.set_fontsize(14)   
        
    for i, eixo in enumerate(valor_inicial):
        txt = f'R$ {eixo:,.2f}'.replace('.', ',').replace('_', '.')
        x_coord = i -0.33
        y_coord = eixo +15
        ax.text(x = x_coord, y = y_coord, s = txt, fontsize = 12)
      
       

window = Tk()

window.title('RAS Investimenros - Renda Fixa')
window.geometry("580x790")

valor_inicial = tk.StringVar()

window.configure(bg = "#ffffff")
canvas = Canvas(
    window,
    bg = "#ffffff",
    height = 790,
    width = 580,
    bd = 0,
    highlightthickness = 0,
    relief = "ridge")
canvas.place(x = 0, y = 0)

background_img = PhotoImage(file = f"background.png")
background = canvas.create_image(
    290.0, 245.0,
    image=background_img)

entry0_img = PhotoImage(file = f"img_textBox0.png")
entry0_bg = canvas.create_image(
    448.5, 211.5,
    image = entry0_img)

entry0 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry0.place(
    x = 362.0, y = 200,
    width = 173.0,
    height = 21)

entry1_img = PhotoImage(file = f"img_textBox1.png")
entry1_bg = canvas.create_image(
    509.0, 298.5,
    image = entry1_img)

entry1 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry1.place(
    x = 483.0, y = 287,
    width = 52.0,
    height = 21)

entry2_img = PhotoImage(file = f"img_textBox2.png")
entry2_bg = canvas.create_image(
    448.5, 72.5,
    image = entry2_img)

entry2 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry2.place(
    x = 362.0, y = 61,
    width = 173.0,
    height = 21)

entry3_img = PhotoImage(file = f"img_textBox3.png")
entry3_bg = canvas.create_image(
    448.5, 105.5,
    image = entry3_img)

entry3 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry3.place(
    x = 362.0, y = 94,
    width = 173.0,
    height = 21)

entry4_img = PhotoImage(file = f"img_textBox4.png")
entry4_bg = canvas.create_image(
    448.5, 138.5,
    image = entry4_img)

entry4 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry4.place(
    x = 362.0, y = 127,
    width = 173.0,
    height = 21)

img0 = PhotoImage(file = f"img0.png")
b0 = Button(
    window,
    image = img0,
    borderwidth = 0,
    highlightthickness = 0,
    command = calc_invest,
    relief = "flat")

b0.place(
    x = 448, y = 163,
    width = 90,
    height = 24)

img1 = PhotoImage(file = f"img1.png")
b1 = Button(
    image = img1,
    borderwidth = 0,
    highlightthickness = 0,
    command = conv_juros,
    relief = "flat")

b1.place(
    x = 360, y = 287,
    width = 113,
    height = 23)

img2 = PhotoImage(file = f"img2.png")
b2 = Button(
    image = img2,
    borderwidth = 0,
    highlightthickness = 0,
    command = conv_tempo,
    relief = "flat")

b2.place(
    x = 360, y = 341,
    width = 113,
    height = 23)

entry5_img = PhotoImage(file = f"img_textBox5.png")
entry5_bg = canvas.create_image(
    509.0, 352.5,
    image = entry5_img)

entry5 = Entry(
    window,
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry5.place(
    x = 483.0, y = 341,
    width = 52.0,
    height = 21)

entry6_img = PhotoImage(file = f"img_textBox6.png")
entry6_bg = canvas.create_image(
    324.0, 298.5,
    image = entry6_img)

entry6 = Entry(
    window,
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry6.place(
    x = 298.0, y = 288,
    width = 52.0,
    height = 21)

entry7_img = PhotoImage(file = f"img_textBox7.png")
entry7_bg = canvas.create_image(
    324.0, 352.5,
    image = entry7_img)

entry7 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry7.place(
    x = 298.0, y = 342,
    width = 52.0,
    height = 21)

img3 = PhotoImage(file = f"img3.png")
b3 = Button(
    image = img3,
    borderwidth = 0,
    highlightthickness = 0,
    command = conv_cdi_cdb,
    relief = "flat")

b3.place(
    x = 342, y = 437,
    width = 113,
    height = 23)

entry8_img = PhotoImage(file = f"img_textBox8.png")
entry8_bg = canvas.create_image(
    165.0, 448.5,
    image = entry8_img)

entry8 = Entry(
    window,
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry8.place(
    x = 132.0, y = 437,
    width = 66.0,
    height = 21)

entry9_img = PhotoImage(file = f"img_textBox9.png")
entry9_bg = canvas.create_image(
    282.0, 448.5,
    image = entry9_img)

entry9 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry9.place(
    x = 232.0, y = 437,
    width = 100.0,
    height = 21)

entry10_img = PhotoImage(file = f"img_textBox10.png")
entry10_bg = canvas.create_image(
    500.0, 448.5,
    image = entry10_img)

entry10 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry10.place(
    x = 465.0, y = 437,
    width = 70.0,
    height = 21)

entry11_img = PhotoImage(file = f"img_textBox11.png")
entry11_bg = canvas.create_image(
    448.5, 39.5,
    image = entry11_img)

entry11 = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

entry11.place(
    x = 362.0, y = 27,
    width = 173.0,
    height = 23)

window.resizable(False, False)
window.mainloop()
```
